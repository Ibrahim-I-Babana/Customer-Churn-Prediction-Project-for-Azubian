### **Introduction** ###

**General**

*Machine learning allows the user to feed a computer algorithm an immense amount of data and have the computer analyze and make data-driven recommendations and decisions based on only the input data.*
*In most of the situations we want to have a machine learning system to make **predictions**, so we have several categories of machine learning tasks depending on the type of prediction needed: **Classification, Regression, Clustering, Generation**, etc.*

**Classification:** *It is the task whose goal is the prediction of the label of the class to which the input belongs (e.g., Classification of images in two classes: cats and dogs).*

**Regression:** *It is the task whose goal is the prediction of numerical value(s) related to the input (e.g., House rent prediction, Estimated time of arrival ).*

**Generation:** *Itis the task whose goal is the creation of something new related to the input (e.g., Text translation, Audio beat generation, Image denoising ). **Clustering** is the task of grouping a set of objects in such a way that objects in the same group (called a **cluster**) are more similar (in some sense) to each other than to those in other **clusters** (e.g., Clients clutering).*

*In machine learning, there are learning paradigms that relate to one aspect of the dataset: **the presence of the label to be predicted**.* 

**Supervised Learning:** *It is the paradigm of learning that is applied when the dataset has the label variables to be predicted, known as ` y variables`.* 

**Unsupervised Learning:** *It is the paradigm of learning that is applied when the dataset has not the label variables to be predicted.*

**Self-supervised Learning:** *It is the paradigm of learning that is applied when part of the X dataset is considere as the label to be predicted (e.g., the Dataset is made of texts and the model try to predict the next word of each sentence).*

## Hypothesis analysis

## Hypothesis

**Null hypothesis(H0)**: There is no significant relationship between the customers' characteristics and the churn rate. In other words, the variables in the dataset have no impact on customer churn.

**Alternative hypothesis(H1)**: There is a significant relationship between the customers' characteristics and the churn rate. The variables in the dataset have an impact on customer churn.

```python
#testing hypothesis on categorical columns
categorical_columns=['Region', 'Tenure','MRG', 'Top_Pack' ]

#significance level
alpha = 0.05

for column in categorical_columns:
    contingency_table = pd.crosstab(Train_data[column], Train_data['Churn'])

    # Performing chi-square test
    chi2_stat, p_value, _, _ = chi2_contingency(contingency_table)
    print(p_value)

    # Testing the hypothesis

    if p_value < alpha:


        print(f"{column}: There is a significant relationship with churn (reject null hypothesis)")
    else:
        print(f"{column}: There is no significant relationship with churn (fail to reject null hypothesis)")
```

### Create Pipelines

```python
#logTransformer class
class LogTransformer():
    def __init__(self, constant=1e-5):
        self.constant = constant

    def fit(self, x, y=None):
        return self
    def transform (self,x):
        return np.log1p(x + self.constant)

#Create a pipeline for numeric columns
numeric_pipelines = Pipeline(steps=[
    ('num_imputer', SimpleImputer(strategy='mean')),
    ('log_transformation', FunctionTransformer(LogTransformer().transform)),
    ('scaler', RobustScaler()),  
])


#categorical columns pipelines
categorical_pipelines = Pipeline(steps=[
    ('cat_imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(handle_unknown = 'ignore')),
])


 
#Combining preprocessors 
preprocessor = ColumnTransformer([
    ('numeric_pipelines',numeric_pipelines,numeric_features),
    ('categorical_pipelines',categorical_pipelines,categorical_features)
])
```

### Label Encoding

```python
Label_encoder = LabelEncoder()
y_train_encoded = Label_encoder.fit_transform(y_train)
y_test_encoded = Label_encoder.transform(y_test)
```

### Machine Learning on Unbalanced Dataset

#### Gradient Boosting Model

```python
#Creating a gradient boosting model
Gradient_B_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', GradientBoostingClassifier())
])

#Training/Fitting the model
Gradient_B_pipeline.fit(X_train,y_train_encoded)
```

```python
#Evaluating the model
GB_eval_predict = Gradient_B_pipeline.predict(X_test)

#Generating a classification report
GB_report = classification_report(y_test_encoded,GB_eval_predict)
print("Classification Report:\n", GB_report)
```

```python
# If GB_eval_predict contains predicted labels, you can skip this step
# Define model predictions for the model and true labels
GB_probabilities = Gradient_B_pipeline.predict_proba(X_test)[:, 1]

# Compute the AUC score
GB_auc_score = roc_auc_score(y_test_encoded, GB_probabilities)

print("AUC Score:", GB_auc_score)
```

#### Logistic regression Model

```python
#Creating a Logistic Regression model

LR_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(random_state=42)
    )
])

#Training/Fitting 
LR_pipeline.fit(X_train,y_train_encoded)
```

```python
#Define model predictions for the model and true labels
LR_probabilities = LR_pipeline.predict_proba(X_test)[:, 1]

# Compute the AUC score
LR_auc_score = roc_auc_score(y_test_encoded, LR_probabilities)

print("AUC Score:", LR_auc_score)
```

#### Decision Tree Model
